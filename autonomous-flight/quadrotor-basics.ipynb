# Quadrotor Basics

## Rotors
A single spinning rotor has angular momentum that creates a torque on the mass it's attached too, causing it to spin in the opposite direction. With single rotor helicopters, we use a tail rotor to counteract this effect and keep the helicopter balanced. The tail rotor is mounted at 90 degrees to the main propeller, exerting a torque in the opposite direction of the main rotor.

This means, when using propellers we need at least two of them, both spinning in opposite directions to allow aircraft to lift and move.

Quadrotors have four rotors, two that spin in one direction, and two in the other (both powered by separate motors). Normally, quadrotors have an X shape frame (body), where the rotors opposite to each other spin in the same direction. When all four spin at the same time, the vehicle has a net torque of 0. 

## Torque

A measurement of the force which causes something to rotate around a point (turning force).

$$
\tau = r \cdot F \cdot \sin(\theta)
$$

Where:

- $\tau$ - Torque (Newton-meters, Nm)
- $r$ - the distance from the pivot point (length) to where the force is applied (meters, m)
- $F$ - Magnitude of the force applied (Newtons, N)
- $\theta$ - the angle between direction and the line from the pivot point to the application point of the force

If the force is perpendicular ($\theta = 90^\circ$), then $\sin(90) = 1$. This gives us a simplified equation of:

$$
\tau = r \cdot F
$$

The more force applied and the longer the distance from the pivot, the easier it is to create rotation.

Imagine a door that is 0.75 meters wide and has 3 forces applied to it, all with the same force (15 Newtons) that acts perpendicularly to the door:

- 0.25m - from the hinge
- 0.5m - from the hinge (center of the door)
- 0.75m - from the hinge (edge of the door)

We can calculate this in Python code like so:

```python
from display import plot_door

force = 15  # newtons
distances = [0.25, 0.5, 0.75]

torques = [r * force for r in distances]
plot_door(distances, torques)
```
